# Data Cleaning

In this section, you will need to read the CSV files into DataFrames and perform any necessary data cleaning steps. After cleaning, combine all DataFrames into a single DataFrame.

Files:

* `whale_returns.csv`: Contains returns of some famous "whale" investors' portfolios.

* `algo_returns.csv`: Contains returns from the in-house trading algorithms from Harold's company.

* `sp500_history.csv`: Contains historical closing prices of the S&P 500 Index.

## Whale Returns

Read the Whale Portfolio daily returns and clean the data

```python
# Reading whale returns
whale_data = Path("Resources/whale_returns.csv")
whale_df = pd.read_csv(whale_data, index_col='Date', infer_datetime_format=True, parse_dates=True)
```

```python
# Count nulls
whale_df.isnull().sum()
```

```python
# Drop nulls
whale_df = whale_df.dropna()
whale_df.isnull().sum()
```

## Algorithmic Daily Returns

Read the algorithmic daily returns and clean the data

```python
# Reading algorithmic returns
algo_data = Path("Resources/algo_returns.csv")
algo_df = pd.read_csv(algo_data, index_col='Date', infer_datetime_format=True, parse_dates=True)
```

```python
# Count nulls
algo_df.isnull().sum()
```

```python
# Drop nulls
algo_df = algo_df.dropna()
algo_df.isnull().sum()
```

## S&P 500 Returns

Read the S&P 500 historic closing prices and create a new daily returns DataFrame from the data. 

```python
# Reading S&P 500 Closing Prices
sp500_data = Path("Resources/sp500_history.csv")
sp500_df = pd.read_csv(sp500_data, index_col='Date', infer_datetime_format=True, parse_dates=True)
sp500_df.sort_index(inplace=True)
```

```python
# Check Data Types
sp500_df.dtypes
```

```python
# Fix Data Types
sp500_df["Close"] = sp500_df["Close"].str.replace("$" , "")
sp500_df["Close"] = sp500_df["Close"].str.replace("," , "")
sp500_df["Close"] = sp500_df["Close"].astype(float)
sp500_df
```

```python
# Calculate Daily Returns
sp500_dr = sp500_df.pct_change()
sp500_dr
```

```python
# Drop nulls
sp500_dr = sp500_dr.dropna()
sp500_dr.isnull().sum()
```

```python
# Rename `Close` Column to be specific to this portfolio.
columns = ['SP500']
sp500_dr.columns = columns
sp500_dr.head()
```

## Combine Whale, Algorithmic, and S&P 500 Returns

```python
# Join Whale Returns, Algorithmic Returns, and the S&P 500 Returns into a single DataFrame with columns for each portfolio's returns.
portfolio_col = pd.concat([sp500_dr, algo_df, whale_df], axis="columns", join="inner")
portfolio_col
```

## Risk Analysis

Determine the _risk_ of each portfolio:

1. Create a box plot for each portfolio. 
2. Calculate the standard deviation for all portfolios
4. Determine which portfolios are riskier than the S&P 500
5. Calculate the Annualized Standard Deviation

### Create a box plot for each portfolio


```python
# Box plot to visually show risk
portfolio_col.plot.box(figsize=(20,10))
```

### Calculate Standard Deviations

```python
# Calculate the daily standard deviations of all portfolios
portfolio_std = portfolio_col.std()
portfolio_std.head()
```

### Determine which portfolios are riskier than the S&P 500

```python
# Calculate  the daily standard deviation of S&P 500
sp500_std = portfolio_col['SP500'].std()

# Determine which portfolios are riskier than the S&P 500
portfolio_col.std() > sp500_std
```

### Calculate the Annualized Standard Deviation

```python
# Calculate the annualized standard deviation (252 trading days)
portfolio_an_std = portfolio_std * np.sqrt(252)
portfolio_an_std.head()
```

## Rolling Statistics

Risk changes over time. Analyze the rolling statistics for Risk and Beta. 

1. Calculate and plot the rolling standard deviation for the S&P 500 using a 21-day window
2. Calculate the correlation between each stock to determine which portfolios may mimick the S&P 500
3. Choose one portfolio, then calculate and plot the 60-day rolling beta between it and the S&P 500

### Calculate and plot rolling `std` for all portfolios with 21-day window

```python
# Calculate the rolling standard deviation for all portfolios using a 21-day window
portfolio_std21 = portfolio_col.rolling(window=21).std().dropna()
portfolio_std21.head()

# Plot the rolling standard deviation
portfolio_std21.plot(figsize=(20,10), title='Rolling 21-Day Standard Deviations for the Portfolio Analysis')
```

### Calculate and plot the correlation

```python
# Calculate the correlation
portfolio_cor = portfolio_col.corr()

# Display de correlation matrix
portfolio_cor
```

### Calculate and Plot Beta for a chosen portfolio and the S&P 500

```python
# Calculate covariance of a single portfolio
covariance_algo1 = portfolio_col['Algo 1'].cov(portfolio_col['SP500'])
#covariance_algo1

# Calculate variance of S&P 500
variance_sp500 = portfolio_col['SP500'].var()
#variance_sp500

# Computing beta
algo1_beta = covariance_algo1 / variance_sp500
algo1_beta
```

```python
# Plot beta trend
algo1_rolling_covariance = portfolio_col['Algo 1'].rolling(window=30).cov(portfolio_col['SP500'])
rolling_variance_sp500 = portfolio_col['SP500'].rolling(window=30).var()
algo1_rolling_beta = algo1_rolling_covariance / rolling_variance_sp500
algo1_rolling_beta.plot(figsize=(20, 10), title='Rolling 30-DAy Beta of Algo 1')
```

## Rolling Statistics Challenge: Exponentially Weighted Average 

An alternative way to calculate a rolling window is to take the exponentially weighted moving average. This is like a moving window average, but it assigns greater importance to more recent observations. Try calculating the [`ewm`](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.ewm.html) with a 21-day half-life.

```python
# Use `ewm` to calculate the rolling window
portfolio_ewm = portfolio_col.ewm(halflife=21).std().dropna()
portfolio_ewm.plot(figsize=(20,10), title='Exponentially Weighted Average with a 21-Day half-life')
```

# Sharpe Ratios
In reality, investment managers and thier institutional investors look at the ratio of return-to-risk, and not just returns alone. After all, if you could invest in one of two portfolios, and each offered the same 10% return, yet one offered lower risk, you'd take that one, right?

### Using the daily returns, calculate and visualize the Sharpe ratios using a bar plot

```python
# Annualized Sharpe Ratios
sharpe_ratios_port = (portfolio_col.mean() * 252) / (portfolio_col.std() * np.sqrt(252))
sharpe_ratios_port
```

```python
# Visualize the sharpe ratios as a bar plot
sharpe_ratios_port.plot.bar(figsize=(20,10), title='Sharp Ratios annualized for the Portfolio analysis')
```

## Choose 3-5 custom stocks with at last 1 year's worth of historic prices and create a DataFrame of the closing prices and dates for each stock.

For this demo solution, we fetch data from three companies listes in the S&P 500 index.

* `GOOG` - [Google, LLC](https://en.wikipedia.org/wiki/Google)

* `AAPL` - [Apple Inc.](https://en.wikipedia.org/wiki/Apple_Inc.)

* `COST` - [Costco Wholesale Corporation](https://en.wikipedia.org/wiki/Costco)

```python
#Downloading stock infromatino from Yahoo Finance
aapl_df2 = yf.download("aapl", start='2015-03-03', end="2019-04-23")
tsla_df2 = yf.download("tsla", start='2015-03-03', end="2019-04-23")
amzn_df2 = yf.download("amzn", start='2015-03-03', end="2019-04-23")
msft_df2 = yf.download("msft", start='2015-03-03', end="2019-04-23")
goog_df2 = yf.download("goog", start='2015-03-03', end="2019-04-23")

#Dropping extra columns and renaming close column to be able to combine data into a clean data frame for analysis
aapl_df2 = aapl_df2.drop(columns = ['Open', 'High', 'Low', 'Adj Close', 'Volume']).rename(columns = {'Close':'AAPL'})
tsla_df2 = tsla_df2.drop(columns = ['Open', 'High', 'Low', 'Adj Close', 'Volume']).rename(columns = {'Close':'TSLA'})
amzn_df2 = amzn_df2.drop(columns = ['Open', 'High', 'Low', 'Adj Close', 'Volume']).rename(columns = {'Close':'AMZN'})
msft_df2 = msft_df2.drop(columns = ['Open', 'High', 'Low', 'Adj Close', 'Volume']).rename(columns = {'Close':'MSFT'})
goog_df2 = goog_df2.drop(columns = ['Open', 'High', 'Low', 'Adj Close', 'Volume']).rename(columns = {'Close':'GOOG'})
```

```python
# Combine all stocks in a single DataFrame
portfolio_cust = pd.concat([aapl_df2, amzn_df2, tsla_df2, msft_df2, goog_df2], axis="columns", join="inner")
portfolio_cust.sort_index(inplace=True)
portfolio_cust
```

```python
# Reset Date index
portfolio_cust.plot(figsize=(20,10))
```

```python
# Calculate daily returns and drop NA's
portfolio_dr = portfolio_cust.pct_change().dropna()

# Display sample data
portfolio_dr.head()
```

## Calculate the weighted returns for the portfolio assuming an equal number of shares for each stock

```python
# Set weights
weights = [.2, .4, .2, .1, .1]

# Calculate portfolio return
portfolio_returns = portfolio_dr.dot(weights)

# Display sample data
new_portfolio_returns_df = pd.DataFrame(portfolio_returns).rename(columns = {0:'PORT1'})
new_portfolio_returns_df.sort_index(inplace=True)
new_portfolio_returns_df
```

## Join your portfolio returns to the DataFrame that contains all of the portfolio returns

```python
# Join your returns DataFrame to the original returns DataFrame
combined_port = pd.concat([new_portfolio_returns_df, portfolio_col], axis="columns", join="inner").dropna()
combined_port.head()
```

```python
# Only compare dates where return data exists for all the stocks (drop NaNs)
combined_port.plot(figsize = (20,10))
```

## Re-run the risk analysis with your portfolio to see how it compares to the others

### Calculate the Annualized Standard Deviation

```python
# Calculate the annualized `std`
combined_port_std = combined_port.std()
combined_port_std.head()

combined_port_an_std = combined_port_std * np.sqrt(252)
combined_port_an_std.head()
```

### Calculate and plot rolling `std` with 21-day window

```python
# Calculate rolling standard deviation
combined_std21 = combined_port.rolling(window=21).std().dropna()

# Plot the rolling standard deviation
combined_std21.plot(figsize=(20,10), title='Rolling 21-Day Standard Deviation for each Portfolio')

```

### Calculate and plot the correlation

```python
# Calculate the correlation
combined_cor = combined_port.corr()

# Display de correlation matrix
combined_cor
```

### Calculate and Plot Rolling 60-day Beta for Your Portfolio compared to the S&P 500

```python
# Calculate covariance of a the custum portfolio
covariance_port1 = combined_port['PORT1'].cov(combined_port['SP500'])

# Calculate variance of S&P 500
variance_sp500 = combined_port['SP500'].var()

# Computing beta
port1_beta = covariance_port1 / variance_sp500
port1_beta
```

```python
# Plot beta trend
port1_rolling_covariance = combined_port['PORT1'].rolling(window=60).cov(combined_port['SP500'])
sp500_rolling_variance = combined_port['SP500'].rolling(window=60).var()
port1_rolling_beta = port1_rolling_covariance / sp500_rolling_variance
port1_rolling_beta.plot(figsize=(20, 10), title='Rolling 60-Day Beta of Custom Portolio')
```

### Using the daily returns, calculate and visualize the Sharpe ratios using a bar plot

```python
# Calculate Annualzied Sharpe Ratios
sharpe_ratios = (combined_port.mean() * 252) / (combined_port.std() * np.sqrt(252))
sharpe_ratios
```

```python
# Visualize the sharpe ratios as a bar plot
sharpe_ratios.plot.bar(figsize = (20,10))
```
